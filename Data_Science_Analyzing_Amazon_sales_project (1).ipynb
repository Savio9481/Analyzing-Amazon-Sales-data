```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Amazon Sales data.csv")
data
```

```python
data.shape
```

```python
data.info()
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data
```

```python
data.info()
```

```python
data['shipping duration']=(data['Ship Date']-data['Order Date'])/np.timedelta64(1,'D')
data['ship day name']=(data['Ship Date']).dt.day_name()
```

```python
data.head()
```

```python
data['Unit Profit']=round(data['Unit Price']-data['Unit Cost'],2)
data
```

```python
data.isnull().sum(axis=0).reset_index()
```

```python
def order_priority(x):
  if x=='H':
    return 'High'
  elif x=='C':
    return 'Critical'
  elif x=='L':
    return 'Low'
  elif x=='M':
    return 'Medium'

data['Order Priority']=data['Order Priority'].apply(order_priority)
```

```python
data.head()
```

```python
data.to_csv('/content/sample_data/sales_records_cleaned.csv',index = False)
```

Data Analysis and Visualisation

```python
sales_data=pd.read_csv('/content/sample_data/sales_records_cleaned.csv')
sales_data
```

```python
continuous_columns = ['Unit Price','Unit Cost','Total Revenue','Total Cost','Total Profit','Unit Profit','Units Sold', 'shipping duration']
```

```python
sales_data[continuous_columns].head()
```

```python
for feature in continuous_columns:
  import warnings
  warnings.filterwarnings('ignore')
  fig,ax=plt.subplots(1,2,figsize=(12,6))
  ax[0].hist(sales_data[feature])
  ax[0].set_title('histogram of column values in ' +feature)
  sns.boxplot(sales_data[feature],ax=ax[1])
  ax[1].set_title('boxplot of column values in' +feature)
  sns.set(style='ticks', palette='deep', font_scale=1.1)
  plt.show()
```

**Bivariate Analysis**

Total Profit vs Month

```python
df_month = sales_data.groupby(['Month'])


df_month = df_month.sum().reset_index()

df_month
```

```python
df_month = df_month.sort_values('Month', key = lambda x : pd.to_datetime(x, format='%b').dt.month)

df_month
```

```python
fig, ax = plt.subplots(figsize=(15,7))

sns.lineplot(x = df_month['Month'],y = df_month['Total Profit'], data = df_month, ax = ax)

ax.set_title('Line Graph for Total Profit vs Month')

plt.show()
```

Units Sold vs Month

```python
fig, ax = plt.subplots(figsize=(15,7))

sns.lineplot(x = df_month['Month'],y = df_month['Units Sold'], data = df_month, ax = ax)

ax.set_title('Line Graph for Total Units Sold vs Month')

plt.show()
```
